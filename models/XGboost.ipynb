# Feature engineering
* Split data into year month day and isweekend because we need use date to create new features that is related to our sales.

```python
merge_train_df.head()
```

```python
merge_test_df.isnull().sum()
```

```python
def split(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Year'] = df.Date.dt.year
    df['Month'] = df.Date.dt.month
    df['Day'] = df.Date.dt.day
    df['WeekOfYear'] = df.Date.dt.isocalendar().week
    df['IsWeekend'] = (df['DayOfWeek'] >= 5).astype(int)
```

```python
split(merge_train_df)
split(merge_test_df)
```

```python
merge_train_df.sample(100)
```

```python
# we only choose store is open since if store is close, there is no sale
merge_train_df = merge_train_df[merge_train_df.Open == 1].copy()
```

```python
merge_train_df.columns
```

```python
def competition_month(df):
    '''we want to know how long has the store been open'''
    df['CompetitionOpen'] = 12*(df.Year-df.CompetitionOpenSinceYear) + (df.Month - df.CompetitionOpenSinceMonth)
    df['CompetitionOpen'] = df['CompetitionOpen'].map(lambda x: 0 if x < 0 else x).fillna(0)

competition_month(merge_train_df)
competition_month(merge_test_df)
```

# OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

eod = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
eod.fit(X_input[categorical])
```

```python
new_columns = eod.get_feature_names_out(categorical)
new_columns
```

```python
# transform dataset
encoded_dense_train = eod.transform(X_input[categorical])
encoded_dense_test = eod.transform(X_input_test[categorical])

# Create dataframe 
encoded_train_input = pd.DataFrame(
    encoded_dense_train,
    columns=new_columns,
    index = X_input.index
)

encoded_test_input = pd.DataFrame(
    encoded_dense_test,
    columns=new_columns,
    index = X_input_test.index
)

X_input.drop(labels=categorical,axis=1,inplace=True)
X_input_test.drop(labels=categorical,axis=1,inplace=True)

X_input = pd.concat([X_input,encoded_train_input], axis=1)
X_input_test = pd.concat([X_input_test, encoded_test_input], axis=1)
```

# Model

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_input,y)
```

```python
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_input_test.shape
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
model = XGBRegressor(n_estimator = 100, learning_rate = 0.01)
model.fit(X_train,y_train)
```

```python
rmse = np.sqrt(mean_squared_error(y_val,model.predict(X_val)))
rmse
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200, 300]
}
model = XGBRegressor()
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=3)
grid_search.fit(X_train, y_train)
print(grid_search.best_params_)
```

```python
model = XGBRegressor(n_estimator = 300, learning_rate = 0.2,max_depth=7,random_state=42)
model.fit(X_train,y_train)
```

```python
rmse = np.sqrt(mean_squared_error(y_val,model.predict(X_val)))
rmse
```

# RandomSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform
param_dist = {
    'max_depth': randint(3, 10),
    'learning_rate': uniform(0.01, 0.2),
    'n_estimators': randint(100, 500)
}

model_1 = XGBRegressor()
random_search = RandomizedSearchCV(estimator=model_1, param_distributions=param_dist, n_iter=50, scoring='neg_mean_squared_error', cv=3)
random_search.fit(X_train, y_train)
print(random_search.best_params_)
```

```python
model1 = XGBRegressor(n_estimator = 348, learning_rate = 0.17,max_depth=8,random_state=42)
model1.fit(X_train,y_train)
rmse = np.sqrt(mean_squared_error(y_val,model1.predict(X_val)))
print(rmse)
```

# Bayesian Optimization

```python
import optuna

def objective(trial):
    param = {
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2),
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
    }
    model = XGBRegressor(**param)
    model.fit(X_train, y_train)
    preds = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, preds))
    return rmse

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)
print(study.best_params_)
```
